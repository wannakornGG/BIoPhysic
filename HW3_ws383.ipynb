**Problem 2 - Implement the Gillespie Algorithm for a birth-death process (6pts)**

We want to simulated the following molecular system using the Gillespie
algorithm: 
\begin{cases}
\frac{dR}{dt} = k - s R\\
\frac{dP}{dt} = 4R - \gamma P
\end{cases}

This describes the production of mRNA and the synthesis of 4 proteins per mRNA transcript. Use $k$ = 1/min, $s$ = 1/min, and $\gamma$ = 0.025/min.


**(a)** Write a function that simulates this system using the Gillespie Algorithm. You may start from the code that was presented in class and modify it as needed.

**(b)** Write a second function that plots the number of $R$ anf $P$ molecules over time.

**(c)** Generate multiple trajectories $R$ versus $t$ and $P$ versus $t$ for
$k=1, s=1$ and $\gamma = 0.025$. 

**(d)** Compare this with the
analytical solutions $R(t) = \frac{k}{s}\left(1-e^{-st}\right)$ and
$P(t) = \frac{4k}{s-\gamma}\left[\frac{1}{\gamma}\left(1- e^{-\gamma t}\right)-\frac{1}{s}\left(1- e^{-s t}\right)\right]$.





```python
#Problem 2 a. a function that simulate R molecule
import numpy as np
k=1
s=1
def Gillespie(Nini, T=1000,params=[k, s]):    
    k = params[0]
    s = params[1]
    N = Nini
    totalTime = 0
    Tarray = [0]
    Narray=[Nini]
    while totalTime < T:  
        R = k + s*N    
        xi = random.rand() ##xi is random number between o and 1
        deltaT = -1/R * log(xi)   
        nu =random.rand() #random between (0,1)
              # birh         death
           # |---------|-------------|
           # 0         k/R.          1
        if nu < k/R: #Birth event  
            N += 1 #N=N+1
            
            Narray.append(N)

        elif nu >= k/R: #death event    
            N -= 1  
            
            Narray.append(N)

        totalTime += deltaT
        
        Tarray.append(totalTime) 
        
    return Tarray, Narray
```
